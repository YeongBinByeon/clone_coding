## 디지털 필터 출력
- 동일한 패턴일 때 가장 큰 값을 얻는다

```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
matplotlib.rcParams['figure.figsize'] = (6,1)
```

# Input Signal

```python
x1 = np.arange(200)
y1 = np.sin(2*np.pi*x1/25)
y2 = np.sin(2*np.pi*x1/50)
y = np.concatenate((y1,y2,y1,y2))
# plt.figure(figsize=(6,1))
plt.plot(y)
```

# Two filters

```python
filter1 = np.concatenate((y1,y1,y1,y1))
filter2 = np.concatenate((y2,y2,y2,y2))
plt.plot(filter1)
plt.show()
plt.plot(filter2)
```

# Filtering

```python
z1 = y*filter1
z2 = y*filter2
```

## Average of Output (Detection)

```python
plt.plot(z1)
w = []
for i in range(4):
    output = z1[i*200:(i+1)*200].mean().round(1)
    w.append(np.repeat(output,200))
plt.plot(np.array(w).ravel(), c='r')
```

```python
plt.plot(z2)
w = []
for i in range(4):
    output = z2[i*200:(i+1)*200].mean().round(1)
    w.append(np.repeat(output,200))
plt.plot(np.array(w).ravel(), c='r')
```
