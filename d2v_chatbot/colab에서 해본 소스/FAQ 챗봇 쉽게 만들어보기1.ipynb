```
import jpype
from konlpy.tag import Kkma

# p = jpype.getDefaultJVMPath()

# jpype.startJVM( p )
```

```
faqs = [["1", "당해년도 납입액은 수정 가능 한가요?", "네, 당해년도 납입액은 12464 화면 등록전까지 수정 가능합니다."],
        ["2", "대리인통보 대상계좌 기준은 어떻게 되나요?", "모계좌 기준 가장 최근에 개설된 계좌의 관리점에서 조회 됩니다."],
        ["3", "등록가능 단말기수는 어떻게 되나요?", "5대까지 등록 가능입니다."],
        ["4", "모바일계좌개설 가능한 시간은 어떻게 되나요?", "08:00 ~ 20:00(영업일만 가능)"],
        ["5", "미국인일때 미국납세자등록번호 작성 방법은 어떻게 되나요??", "계좌주가 미국인일 떼 계좌주의 미국납세자등록번호"],]
```

```


# error : Java package 'kr.lucypark.kkma' is not valid
# resolve : colab 런타임 재시작 -  https://github.com/konlpy/konlpy/issues/81

kkma = Kkma()


def tokenizer_kkma(doc):
  jpype.attachThreadToJVM()
  token_doc = [ '/'.join(word) for word in kkma.pos(doc) ]
  return token_doc

print(tokenizer_kkma(faqs[0][1]))



# 아래는 트위터 형태소 분석기
# from konlpy.tag import Twitter

# twitter = Twitter()
# twitter.pos("질문이나 건의사항은 깃헙 이슈 트래커에 남겨주세요")

```

```
# 리스트에서 각 문장부분 토큰화
token_faqs = [ (tokenizer_kkma(row[1]), row[0]) for row in faqs]
token_faqs

# DoceVec에서 사용하는 태그문서형으로 변경
tagged_faqs = [TaggedDocument(d, [c]) for d, c in token_faqs]
tagged_faqs
```

```
import multiprocessing
cores = multiprocessing.cpu_count()
print(cores)

d2v_faqs = doc2vec.Doc2Vec(
    vector_size = 50, # 학습 vector 차원 갯수. 보통 100~150이 저당하다고 한다.
    alpha = 0.025, # learning rate 값
    min_alpha = 0.025, # learning rate 값
    hs=1,
    negative=0,
    dm=0,
    dbow_words=1,
    min_count=0, # 최소 n번 이상 출현된 단어 사용 
    workers = cores,
    seed = 0,
)

d2v_faqs.build_vocab(tagged_faqs) # d2v이 사전을 만듦

for epoch in range(100):
  d2v_faqs.train(tagged_faqs,
                 total_examples = d2v_faqs.corpus_count,
                 epochs = d2v_faqs.epochs,)
  d2v_faqs.alpha -= 0.0025
  d2v_faqs.min_alpha = d2v_faqs.alpha
```

```
# 들어온 질문 벡터화
# 만든 언어 vector 모델을 가지고 입력된 문장의 vector를 표시해 줌
test_string = "당해년도 납입액은 수정 가능 한가요?"
token_test = tokenizer_kkma(test_string)
predict_vector = d2v_faqs.infer_vector(token_test) 
predict_vector # 50차원이라 50개

d2v_faqs.docvecs.most_similar([predict_vector], topn=2)
```
